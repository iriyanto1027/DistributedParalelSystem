## Latihan 1: Koneksi TCP

```python
# Import library socket karena akan menggunakan IPC socket
import socket

# Definisikan alamat IP binding  yang akan digunakan 
IP = "localhost"

# Definisikan port number binding  yang akan digunakan 
PORT = 12345

# Definisikan ukuran buffer untuk mengirimkan pesan
buffer_size = 1024

# Buat socket bertipe TCP
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Lakukan bind
s.bind((IP, PORT))

# Server akan listen menunggu hingga ada koneksi dari client
s.listen(1)
print('Server siap menerima koneksi')

# Lakukan loop forever
while 1:
    # menerima koneksi
    c, addr = s.accept()
    
    # menampilkan koneksi berupa IP dan port client yang terhubung menggunakan print
    print(f'Mendapatkan koneksi dari {addr}')
    
    # menerima data berdasarkan ukuran buffer
    pesan_client = c.recv(buffer_size)
    
    # menampilkan pesan yang diterima oleh server menggunakan print
    print(f'Pesan dari {addr}: {pesan_client.decode()}')
    
    # mengirim kembali data yang diterima dari client kepada client
    pesan = f'Hai {addr}, pesan anda: {pesan_client.decode()}'
    c.send(pesan.encode())

    # tutup koneksi
    c.close()
```

## Latihan 2: Koneksi UDP

```python
# Import library socket karena akan menggunakan IPC socket
import socket

# Definisikan alamat IP bind dari server
IP = 'localhost'

# Definisikan port number untuk bind dari server
PORT = 12345

# Buat socket bertipe UDP
s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

# Lakukan bind
s.bind((IP, PORT))
print('Server siap menerima koneksi')

# loop forever
while True:
    # terima pesan dari client
    pesan, addr = s.recvfrom(1024) # buffer size is 1024 bytes
    
    # menampilkan hasil pesan dari client
    print(f'Pesan dari {addr}: {pesan.decode()}')
```

## Latihan 3: Upload file

```python
# import library socket karena menggunakan IPC socket
import socket

# definisikan IP untuk binding
IP = "localhost"

# definisikan port untuk binding
PORT = 12345

# definisikan ukuran buffer untuk menerima pesan
buffer_size = 1024

# buat socket (TCP)
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# lakukan binding ke IP dan port
s.bind((IP, PORT))

# lakukan listen
s.listen(1)
print('Server siap menerima koneksi')

#  siap menerima koneksi
c, addr = s.accept()
print (f'IP client yang terhubung: {addr}')

# buka/buat file bernama hasil_upload.txt untuk menyimpan hasil dari file yang dikirim server
# masih hardcoded nama file, bertipe byte
f = open("hasil_upload.txt", "w+b")

# loop forever
while 1:
    # terima pesan dari client
    file = c.recv(buffer_size)

    # tulis pesan yang diterima dari client ke file kita (hasil_upload.txt)
    f.write(file)
    
    # berhenti jika sudah tidak ada pesan yang dikirim
    if not file:
        print(f'File diterima dari: {addr}')
        break
    
# tutup file hasil_upload.txt    
f.close()

#tutup socket
s.close()

# tutup koneksi
c.close()
```

## Latihan 4: Download file

```python
# import library socket karena menggunakan IPC socket
import socket

# definisikan IP untuk binding
IP = 'localhost'

# definisikan port untuk binding
PORT = 12345

# definisikan ukuran buffer untuk menerima pesan
buffer_size = 1024

# buat socket (bertipe TCP)
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# lakukan binding ke IP dan port
s.bind((IP, PORT))

# lakukan listen
s.listen(1)
print('Server siap menerima koneksi')

#  siap menerima koneksi
c, addr = s.accept()
print(f'IP Client yang terhubung: {addr}')

# buka file bernama "file_didownload.txt
# masih hard code, file harus ada dalam folder yang sama dengan script python
f = open("file_didownload.txt", "rb")

try:
    # baca file tersebut sebesar buffer 
    byte = f.read(buffer_size)
    
    # selama tidak END OF FILE; pada pyhton EOF adalah b''
    while byte != b'':
        # kirim hasil pembacaan file dari server ke client
        c.send(byte)
        
        # baca sisa file hingga EOF
        byte = f.read(buffer_size)
        
finally:
    print('Pengiriman selesai')
    
    # tutup file jika semua file telah  dibaca
    f.close()

# tutup socket
s.close()

# tutup koneksi
c.close()
```
